# Problem Statement

The objective of this project is to perform exploratory data analysis on the house prediction dataset to understand data distributions, statistical summaries, correlations, and patterns using different visualization techniques.

```python
# Import Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("house Prediction Data Set.csv")

# Display First 5 Rows

df.head()
```

```python
# Display Dataset Shape

df.shape
```

```python
# Display Dataset Information

df.info()
```

```python
# Display Summary Statistics

df.describe()
```

```python
# Calculate Mean

df.mean(numeric_only=True)
```

```python
# Calculate Median

df.median(numeric_only=True)
```

```python
# Calculate Mode

df.mode().head()
```

```python
# Calculate Standard Deviation

df.std(numeric_only=True)
```

```python
# Correlation Matrix

df.corr(numeric_only=True)
```

```python
# Create Heatmap

plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Histogram Plot

df.hist(figsize=(12,10))

plt.show()
```

```python
# Create Boxplot

plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.xticks(rotation=90)

plt.title("Boxplot of Dataset")

plt.show()
```

```python
# Scatter Plot

plt.figure(figsize=(6,4))

sns.scatterplot(x=df.iloc[:,0],
                y=df.iloc[:,1])

plt.title("Scatter Plot")

plt.show()
```

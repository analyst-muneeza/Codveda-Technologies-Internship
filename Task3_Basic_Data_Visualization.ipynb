# Problem Statement

The objective of this project is to create basic data visualizations to understand the distribution and relationships within the house prediction dataset using different plotting techniques.

```python
# Import Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset

df = pd.read_csv("house Prediction Data Set.csv")

# Display First 5 Rows

df.head()
```

```python
# Create Line Chart

plt.figure(figsize=(8,5))

y = df.iloc[:20,5]

plt.plot(y,
         marker='o')

plt.title("Line Chart")

plt.xlabel("Index")

plt.ylabel("Values")

plt.show()
```

```python
# Create Bar Plot

plt.figure(figsize=(8,5))

x = ['A','B','C','D','E']
y = df.iloc[:5,5]

plt.bar(x, y)

plt.title("Bar Plot")

plt.xlabel("Categories")

plt.ylabel("Values")

plt.show()
```

```python
# Create Pie Chart

data = df.iloc[:,3].value_counts()

plt.figure(figsize=(6,6))

plt.pie(data,
        labels=data.index,
        autopct='%1.1f%%')

plt.title("Pie Chart")

plt.show()
```

```python
# Create Countplot

plt.figure(figsize=(8,5))

sns.countplot(x=df.iloc[:,3])

plt.title("Countplot")

plt.show()
```

```python
# Create Scatter Plot

plt.figure(figsize=(8,5))

sns.scatterplot(x=df.iloc[:,5],
                y=df.iloc[:,6])

plt.title("Scatter Plot")

plt.xlabel("Column 6")

plt.ylabel("Column 7")

plt.show()
```

```python
# Create Histogram

df.hist(figsize=(12,10))

plt.show()
```

```python
# Create Boxplot

plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.xticks(rotation=90)

plt.title("Boxplot")

plt.show()
```

```python
# Create Heatmap

plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Create Pairplot

sns.pairplot(df)

plt.show()
```

```python
# Save Scatter Plot

plt.figure(figsize=(6,4))

sns.scatterplot(x=df.iloc[:,0],
                y=df.iloc[:,1])

plt.title("Scatter Plot")

plt.savefig("ScatterPlot.png")

plt.show()
```

# Objective

- Analyze stock price trends over time
- Visualize time-series data
- Apply moving average smoothing
- Understand patterns in stock prices

# Problem Statement

The objective of this project is to analyze stock price data using time series analysis techniques to identify trends and patterns through visualization and moving averages.

```python
# Import Libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load Dataset

df = pd.read_csv("Stock Prices Data Set.csv")

# Display First 5 Rows

df.head()
```

```python
# Display Dataset Information

df.info()
```

```python
# Check Missing Values

df.isnull().sum()
```

```python
# Convert Date Column into Datetime Format

df['date'] = pd.to_datetime(df['date'])

# Display Dataset

df.head()
```

```python
# Set Date Column as Index

df.set_index('date', inplace=True)

# Display Dataset

df.head()
```

```python
# Plot Closing Price

plt.figure(figsize=(12,6))

plt.plot(df['close'])

plt.title("Stock Closing Price Over Time")

plt.xlabel("Date")

plt.ylabel("Closing Price")

plt.show()
```

```python
# Calculate Moving Average

df['Moving_Average'] = df['close'].rolling(window=30).mean()

# Display Dataset

df.tail()
```

```python
# Plot Closing Price and Moving Average

plt.figure(figsize=(12,6))

plt.plot(df['close'], label='closing Price')

plt.plot(df['Moving_Average'],
         color='red',
         label='30 Days Moving Average')

plt.title("Stock Price Moving Average")

plt.xlabel("Date")

plt.ylabel("Price")

plt.legend()

plt.show()
```

```python
# Histogram of Closing Prices

plt.figure(figsize=(8,5))

plt.hist(df['close'], bins=20)

plt.title("Distribution of Closing Prices")

plt.xlabel("Closing Price")

plt.ylabel("Frequency")

plt.show()
```

```python
# Boxplot of Closing Prices

plt.figure(figsize=(6,4))

plt.boxplot(df['close'])

plt.title("Boxplot of Closing Prices")

plt.show()
```

```python
# Save Plot as Image

plt.figure(figsize=(12,6))

plt.plot(df['close'])

plt.title("Stock Price Trend")

plt.savefig("StockPriceTrend.png")

plt.show()
```

```python
# save clean dataset

df.to_csv("Clean_Stock_Prices_Data.csv", index=False)
```

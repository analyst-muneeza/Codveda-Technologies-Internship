# Problem Statement

The objective of this project is to clean and preprocess the raw house prediction dataset by handling missing values, removing duplicate rows, and improving overall data quality for further analysis.

```python
# Import Pandas Library
import pandas as pd
```

```python
# Load Dataset

df = pd.read_csv("house Prediction Data Set.csv")

# Display First 5 Rows

df.head()
```

```python
# Display Dataset Shape

df.shape
```

```python
# Display Dataset Information

df.info()
```

```python
# Check Missing Values

df.isnull().sum()
```

```python
# Remove Duplicate Rows

df = df.drop_duplicates()

print("Duplicate Rows Removed Successfully")
```

```python
# Select Numerical Columns

num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Fill Missing Numerical Values with Mean

for col in num_cols:
    df[col] = df[col].fillna(df[col].mean())

print("Numerical Missing Values Filled")
```

```python
# Select Categorical Columns

cat_cols = df.select_dtypes(include=['object']).columns

# Fill Missing Categorical Values with Mode

for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("Categorical Missing Values Filled")
```

```python
# Display Data Types

df.dtypes
```

```python
# Check Missing Values After Cleaning

df.isnull().sum()
```

```python
# Display Cleaned Dataset

df.head()
```

```python
# Save Cleaned Dataset

df.to_csv("Cleaned_House_Prediction.csv", index=False)

print("Cleaned Dataset Saved Successfully")
```

```python
# Completion Message

print("Data Cleaning and Preprocessing Completed Successfully")
```
